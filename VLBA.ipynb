2.2. Writign script (Example)




import bpy

from math import radians


Create object eg. cube
bpy.ops.mesh.primitive_cube_add()

# Select an object
cube_selected = bpy.context.active_object


# Translate object
cube_selected.location[0] = 10    # [0] is x, [1] is y and [2] is z-axis index


# Parent object is displaced by -10 in x direction and rotated by 60 degrees

cube_selected.location[0] = -10

cube_neg_selected = bpy.context.active_object


cube_neg_selected.rotation_euler[0] += radians(60)
